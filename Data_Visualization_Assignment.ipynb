1.loc: Mean of the distribution (0 in this case)

3.size: Number of random numbers to generate (1000 in this case)

kde=True: Adds a kernel density estimate curve to the histogram, providing a smoother representation of the distribution.

Reasoning:

2.Normal Distribution: Using np.random.normal allows us to generate data that follows a normal (Gaussian) distribution, which is common in many real-world scenarios.

3.KDE: The kernel density estimate helps to visualize the underlying probability density function of the data, making it easier to identify the shape and characteristics of the distribution.

Q.5 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
# 1. Import Liabraries:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Generate Synthetic Data:

# Set the number of samples and features
num_samples = 100
num_features = 5

# Create a random dataset
data = np.random.rand(num_samples, num_features)

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=['Feature_' + str(i) for i in range(1, num_features + 1)])

# Introduce correlation between features (example: Feature_1 and Feature_2)
df['Feature_2'] = df['Feature_1'] * 0.8 + np.random.rand(num_samples) * 0.2
```

```python
# 3. Calculate Correlation Matrix:

correlation_matrix = df.corr()
```

```python
# 4.Visualize with Heatmap:

plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Reasoning:

1.Synthetic Data Generation: We create a random dataset using np.random.rand and then introduce correlation between features by making one feature partially dependent on another. This simulates real-world scenarios where features are often related.

2.Correlation Matrix Calculation: The df.corr() function calculates the Pearson correlation coefficient between all pairs of features in the DataFrame.

3.Heatmap Visualization: Seaborn's heatmap function is used to visualize the correlation matrix.

annot=True displays the correlation values on the heatmap.

cmap='coolwarm' sets the color scheme.

fmt=".2f" formats the correlation values to two decimal places.

```python
import plotly.express as px
import numpy as np
import pandas as pd

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table to aggregate sales by month and day
sales_pivot = pd.pivot_table(df, values='Sales', index='Month', columns='Day')

# Create the heatmap
fig = px.imshow(sales_pivot,
                x=sales_pivot.columns,
                y=sales_pivot.index,
                color_continuous_scale='Viridis')  # You can choose other color scales
fig.update_layout(title='Sales Variation Across Months and Days')
fig.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_trisurf(df['X'], df['Y'], df['Z'], cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of sin(x + y)')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Area of each bubble is proportional to population
areas = df['Population'] * 2  # Adjust the coefficient for bubble size

# Create the bubble chart
plt.figure(figsize=(8, 6))
plt.scatter(df['GDP'], df['Population'], s=areas, alpha=0.7, c='blue')

# Label and title the chart
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart of Country Population vs. GDP')

# Add country names as annotations
for i, row in df.iterrows():
    plt.annotate(row['Country'], (row['GDP'], row['Population']), fontsize=10)

plt.grid(True)
plt.show()
```

Q.2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
num_points = 50
x = np.random.rand(num_points) * 10  # x values between 0 and 10
y = np.random.rand(num_points) * 10  # y values between 0 and 10
sizes = np.random.randint(5, 20, num_points)  # Random sizes between 5 and 20
colors = np.random.choice(['red', 'blue', 'green'], num_points)  # Random colors

# Create a Pandas DataFrame
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(df)

# Output the plot to the notebook
output_notebook()

# Create a Bokeh figure
p = figure(title="Scatter Plot with Varying Sizes and Colors",
           x_axis_label="x", y_axis_label="y")

# Add a scatter renderer with varying sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

# Show the plot
show(p)
```

Q.5 Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper, BasicTicker, ColumnDataSource

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Reshape data for Bokeh
data = {'x': xx.flatten(),
        'y': yy.flatten(),
        'values': data_heatmap.flatten()}
source = ColumnDataSource(data=data)

# Create color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Output the plot to the notebook
output_notebook()

# Create figure
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1),
           x_axis_label="x", y_axis_label="y",
           tools="hover", tooltips=[("x", "@x"), ("y", "@y"), ("value", "@values")])

# Add heatmap rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color={'field': 'values', 'transform': color_mapper},
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(),
                     label_standoff=12, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
